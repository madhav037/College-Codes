### 1) Write a python program to compute distance between Given two objects represented by the tuples (22, 1, 42, 10) and (20, 0, 36, 8): 
(a) Compute the Euclidean distance between the two objects. <BR>
(b) Compute the Manhattan distance between the two objects. <BR>
(c) Compute the Minkowski distance between the two objects, using q = 3. <BR>
(d) Compute the supremum distance between the two objects.<BR>


```python
import math
x = (22,1,42,10)
y = (20,0,36,8)

dist = 0
for i in range(len(x)):
    dist += (x[i] - y[i])**2
    
dist = math.sqrt(dist)
print(dist)
```

```python
x = (22,1,42,10)
y = (20,0,36,8)

dist = 0
for i in range(len(x)):
    dist += (x[i] - y[i])
dist = abs(dist)

print(dist)
```

### 3) Write programs to perform the following tasks of preprocessing.

Equal Width Binning <BR>
Equal Frequency/Depth Binning

```python
import numpy as np
data = [5,10,11,13,15,35,50,55,72,92,204,215]

num_of_bins = 3
df10 = pd.DataFrame(data,columns=["Value"])
bin_edges = np.linspace(df10["Value"].min(),df10['Value'].max(), num_of_bins+1 )

df10['Equal_with_bins'] = pd.cut(df10['Value'],bins=bin_edges, include_lowest=True, labels=["bin 1", "bin 2", "bin 3"])
df10
```

### 4)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
import pandas as pd
df = pd.read_csv('titanic.csv')
new_min = 0
new_max = 1
df['minMax'] = ((df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())) * (new_max - new_min) + new_min  
df
```

```python
power = len(str(int(df['Age'].max())))
df['decimalScaling'] = df['Age'] / (10**(power))
df
```

```python
df['Zindex'] = (df['Age'] - df['Age'].mean()) / df['Age'].std()
df
```
