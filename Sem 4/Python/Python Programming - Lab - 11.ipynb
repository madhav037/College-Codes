### 03) WAP to demonstrate the use of Bar chart.

```python
import random as rn
x = [rn.randint(1,20) for i in range(1,6)]
y = [rn.randint(1,20) for i in range(1,6)]
l = ['c++', 'python', 'java', 'c#', 'javaScript']
c = ['r', 'g', 'b', 'k', 'y']
w = [0.5,0.6,0.8,0.3,0.9]
plt.title('this is title')
plt.bar(x,y,color=c,label=l,width=w)
```

### 04) WAP to demonstrate the use of Histogram.

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import random as rn
x = [rn.randint(1,50) for i in range(1,6)]
y = [rn.randint(1,20) for i in range(1,6)]
l = ['c++', 'python', 'java', 'c#', 'javaScript']
c = ['r', 'g', 'b', 'k', 'y']
plt.title('this is title')
plt.bar(x,y,color=c,label=l,width=2)

for i in range(len(y)):
    plt.text(x[i],y[i],y[i])
plt.show()
```
