### 04) WAP to find the longest word in a file named abc.txt

```python
with open("abc.txt","r") as f:
    fileList = f.read().split()
    print( max(fileList, key = len) )
```

### 06) WAP to implement search function to search specific occurance of word in a given text file.

```python
f = open("abc.txt","r")
word = input("Enter word to search : ")
fileList = f.read().split()
for i in fileList:
    if word == i:
        print(i)
f.close()
```

### 01) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
f = open("primenumbers.txt","w")
for i in range(1,101):
    for j in range(2,i):
        if i % j == 0:
            break
    else:
        f.write(str(i)+"\n")
f.close()        
```

### 02) WAP to merge two files and write it in a new file.

```python
f = open("abc.txt","r")
f2 = open("primenumbers.txt","r")
fnew = open("newFile.txt","w")

for i in f:
    fnew.write(i)
    
fnew.write("\n")

for i in f2:
    fnew.write(i)
    
f.close()
f2.close()
fnew.close()
```

### 03) WAP to encrypt a text file.

```python
filee = open("encrypt.txt","w",encoding="utf-8")
key = int(input("Enter some key integer : "))
with open("abc.txt","r") as f:
    for line in f.readlines():
        for char in line:
            filee.write(chr(ord(char)+key))
            
filee.close()
```

### 04) WAP to decrypt a previously encrypted file.

```python
filee = open("decrypt.txt","w",encoding="utf-8")
key = int(input("Enter some key integer : "))
with open("encrypt.txt","r", encoding="utf-8") as f:
    for line in f.readlines():
        for char in line:
            filee.write(chr(ord(char)-key))
            
filee.close()
```

### 05) WAP to remove a word from text file.

```python
f = open("abc.txt","r")
word = input("Enter word to remove : ")
x =word+"\n"
fileList = f.read().split()
print(fileList)
fileList.remove(x)
fileList.remove(word)
print(fileList)
f.close()    
```
